```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #yaptığımız grafiklerin direkt görünmesini sağlar
```

```python
age_list = [20, 25, 30, 35, 40, 45, 50, 55, 60]
weight_list = [70, 80, 90, 100, 40, 60, 36, 89, 55]
```

```python
age_list_numpy = np.array(age_list) #bu listeyi alıyoruz age_list_içine. numpy dizisi oluşturduk.
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
plt.plot(age_list_numpy, weight_list_numpy, "m") #plot grafikteki çizimi yapıyor. m ona rengini verdi.
plt.xlabel("Age") # x eksenine age ismini verdik.
plt.ylabel("Weight") # y eksenine weight ismini verdik.
plt.title("Age-Weight Graph")
plt.show() #grafiği çalıştırır.
```

```python
my_numpy1 = np.linspace(0,15,20) #linspace 0'dan 15'e kadar 20'şer olarak artarak bir dizi verir.
```

```python
my_numpy1
```

```python
my_numpy2 = my_numpy1 ** 3 #mynmpy1'in üssü 3 olsun
```

```python
my_numpy2
```

```python
plt.plot(my_numpy1, my_numpy2 ,"m-*") #bütün veri noktalarını yıldız olarak gösterir.
# - koysak hepsini - gösterir. + koysak + gösterir.
#hem * var hem de - var.
# -- yaparsak kesikli gösterir.
```

```python
plt.subplot(1,2,1) #çizeceğimiz grafik subplot'tır yani alt grafiktir. 1 sıra 2 kolon ve 1. grafiği çiziyoruz.
plt.plot(my_numpy1, my_numpy2, "m--")
plt.subplot(1,2,2)
plt.plot(my_numpy2, my_numpy1, "g*")
```

```python
my_figure = plt.figure() #figure ile içine plotları saklayabiliyoruz.
type(my_figure) #buna bir boyut vermemiz lazım.
my_axes = my_figure.add_axes([0.1,0.1,2,2]) # figure dizimizin içine add_axes ile boyutlarını ekledik.
# 1, x ve y değerleridir. 2, bunların değer aralığıdır yani boyutudur.
my_axes.plot(my_numpy1, my_numpy2, "m")
my_axes.set_xlabel("X data") # x boyutuna isim verdik.
my_axes.set_ylabel("Y data") # y bouyutuna isim verdik.
my_axes.set_title("Graph line") # graph'a isim verdik. Başlık.
```

```python
#iç içe geçmiş grafikler
my_figure2 = plt.figure() 

my_axes3 = my_figure2.add_axes([0.1,0.1,0.9,0.9]) #boyutlaro 0.9 belirler.
my_axes3.plot(my_numpy1, my_numpy2, "g*-")
my_axes3.set_xlabel("X large data") 
my_axes3.set_ylabel("Y large data") 
my_axes3.set_title(" Large Graph line") 

my_axes2 = my_figure2.add_axes([0.2,0.4,0.3,0.3]) # iç içe için x ve y boyutlarıyla oynarız. yana gitsin diye. küçük olması için de boyutları ayarlarız.
my_axes2.plot(my_numpy1, my_numpy2, "m--")
my_axes2.set_xlabel("X small data") 
my_axes2.set_ylabel("Y small data") 
my_axes2.set_title(" Small Graph line") 
```

```python
(my_fig, my_axe) = plt.subplots() # (2,2) 2'ye 2 dediğimiz zaman ayrı ayrı axis verir.
my_numpy3 = np.linspace(0,15,20) #yeni dizi.
my_numpy4 = my_numpy3 + 10  #yeni dizi
my_axe.plot(my_numpy3, my_numpy4, color="#7a2d2d", alpha =1)
my_axe.plot(my_numpy4, my_numpy3, color="#ef55cf", alpha=0.5) #alpha saydamlığı belirtir.
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color="#253c66", linewidth = 5) #linewidth ile çizginin kalınlığını belirleriz.
new_axe.plot(my_numpy3, my_numpy3 + 4, color="#FF5733", linewidth = 2)
new_axe.plot(my_numpy3, my_numpy3 + 6, color="#b5e946", linewidth = 1.8, linestyle="-.") #çizginin nasıl olacağını belirler. çizgi nokta olur.
new_axe.plot(my_numpy3, my_numpy3 + 8, color="#7346e9", linewidth = 2, linestyle=":")
new_axe.plot(my_numpy3, my_numpy3 + 10, color="#e9ad46", linewidth = 2, marker="o") #marker o yaptığımız zaman o'lar belirgin görünür.
new_axe.plot(my_numpy3, my_numpy3 + 12, color="#315979", linewidth = 2, linestyle="--", marker="+", markersize=10) #markersize ile marker'ları büyüttük.
```

```python
plt.scatter(my_numpy3, my_numpy4) # scatter, data'ları dağıtarak verir.
```
